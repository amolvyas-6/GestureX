```python
import numpy as np
import pandas as pd
import tensorflow as tf
```

```python
df = pd.read_csv('..\data\gestureData.csv')
NUM_CLASSES = 6
y_train = np.array(df.iloc[:, 0])
x_train = np.array(df.iloc[:, 1:])
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(units=20, activation='relu', name='layer1'),
    tf.keras.layers.Dense(units=20, activation='relu', name='layer2'),
    tf.keras.layers.Dense(units=NUM_CLASSES, activation='linear', name='output')
])
```

```python
model.compile(
    loss = tf.keras.losses.SparseCategoricalCrossentropy(reduction='sum', from_logits=True),
    optimizer = tf.keras.optimizers.Adam(learning_rate=0.01),
)
model.fit(x_train, y_train, epochs=200, verbose=1)
```
